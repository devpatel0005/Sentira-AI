```python
import os
import requests
from dotenv import load_dotenv

# Load token from .env
load_dotenv()
HF_API_KEY = os.getenv("HF_API_KEY", "").strip()

API_URL = "https://router.huggingface.co/v1/chat/completions"
MODEL = "meta-llama/Llama-3.1-8B-Instruct"

headers = {
    "Authorization": f"Bearer {HF_API_KEY}",
    "Content-Type": "application/json"
}

def ask_llm(prompt):
    if not HF_API_KEY:
        return "Error: HF_API_KEY is missing in .env file."

    payload = {
        "model": MODEL,
        "messages": [{"role": "user", "content": prompt}],
        "max_tokens": 250
    }

    response = requests.post(API_URL, headers=headers, json=payload, timeout=60)

    try:
        data = response.json()
    except ValueError:
        return f"Error: Non-JSON response ({response.status_code}) -> {response.text[:300]}"

    if response.status_code != 200:
        return f"Error {response.status_code}: {data}"

    return data["choices"][0]["message"]["content"]


# Example
prompt = "A user said: I am the most happiest guy right now. Emotion: Happiness. Give explanation and 3 suggestions."
print(ask_llm(prompt))
```
